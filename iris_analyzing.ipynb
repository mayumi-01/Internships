```python
## KMeans, Dendrogram, DBSCANを比較して、使い方/特徴を考察してください。
```

KMeansはここのデータポイントを最寄りのクラスタ重心に割り当てる。
	<br>その後、そのクラスタ重心をデータポイントの平均に設定する操作を繰り返す。
	<br>線形分離可能なクラスタに適している。
		<br>
		
Dendrogramは凝集的クラスタリング = 個々のデータポイントをそれぞれ個別のクラスタとして開始し、最も類似した2つのクラスタを併合していく。
を可視化するときに用いる。計算コストが高く、大規模データセットには向かない。だがクラスタ数を指定しなくても適切なクラスタ数を見つけてくれる。
DBSCANは凝集的クラスタリングやKMeans法よりも遅いが、大きなデータセットに対しても適用可能である。
また、クラスタ数を指定しなくても適切なクラスタ数を見つけてくれる利点がある。
高密度領域のデータポイントをクラスタとして認識し、比較的空虚な領域で区切る。
高密度領域の中のデータポイント: コアサンプル
eps以内のコアサンプルは同じクラスタに割り当てられるが、eps以内であってもmin_samples以下の場合はノイズ(-1)として扱われる
結果的に分けられるのは
コアポイント、コアポイントからeps以内にあるデータポイント、ノイズの3つ
